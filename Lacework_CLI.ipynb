### **Multiple Profiles**

The Lacework CLI has the ability to configure multiple profiles if you are managing multiple accounts within the platform. After configuring the CLI with multiple profiles, you can run commands against any of your configured accounts with the --profile flag.

The lacework configure command is the fastest way to set up your Lacework CLI installation. The following command reads the downloaded JSON file and sets up a default configuration for the CLI.

```bash
# lacework configure -j /path/to/API_KEY.json
```

```bash
lacework int list
```

```bash
lacework int list --profile default
```

### **Event Insights**

You can pull events directly from your accounts via the Lacework CLI, filter by severity, get detailed information, and open events directly from the CLI in the UI for further investigation.

```bash
lacework events list --severity high --days 5
```

### **Show Event Details**

Either as human readable or machine structured JSON.

```bash
lacework event show 18706 #<event_id>
```

```bash
lacework event show 18706 --json  #<event_id>
```

### **Container Vulnerability Assessments**

When diving into vulnerabilities in your environment, you’ll first want to get an understanding of current container vulnerability assessments available. We can optionally filter the full list down to those that are active and have fixes available.

```bash
lacework vulnerability container list-assessments --active --fixable
```

### **Host Vulnerability Assessments**

Just like the container vulnerability assessment example, the first thing you’ll want to do is list out all the CVEs found within your environment. We can optionally filter the full list down to what CVEs are active and fixable within your environment.

```bash
lacework vulnerability host list-cves --active --fixable
```

Sometimes you need to zero in on a specific CVE and get an understanding of which hosts have that CVE.

```bash
lacework vulnerability host list-hosts CVE-2020-12049 #<cve_id>
```

### **Compliance Assessments**

The Lacework CLI also has the ability to both retrieve and run compliance assessments of integrated AWS, GCP, and Azure Cloud environments. 

```bash
lacework compliance aws list-accounts
```

```bash
lacework compliance aws get 950194951070 #<account_id>
```

The Lacework CLI can fetch different types of compliance reports and return them in PDF format.

```bash
lacework compliance aws get --pdf --type PCI 463783698038 #<account_id>
```
